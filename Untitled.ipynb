```python
import cv2

import matplotlib.pyplot as plt
import numpy as np
frame = cv2.imread(r'./img/Image 1_2_Channel Alignment.tif')
hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

# Green color
low_green = np.array([55, 50, 0])
high_green = np.array([72, 255, 255])
green_mask = cv2.inRange(hsv_frame, low_green, high_green)
green = cv2.bitwise_and(frame, frame, mask=green_mask)

# Red color
low_red = np.array([100, 0, 0])
high_red = np.array([140, 255, 255])
red_mask = cv2.inRange(hsv_frame, low_red, high_red)
red = cv2.bitwise_and(frame, frame, mask=red_mask)


#blur = cv2.GaussianBlur(nemo, (71, 71), 0)
plt.rcParams['figure.figsize'] = [10, 10]

summed = np.sum(green, axis=2)
tot_pixel = summed.size
green_pixel = np.count_nonzero(summed)
green_percentage = round(green_pixel * 100 / tot_pixel, 2)

summed = np.sum(red, axis=2)
red_pixel = np.count_nonzero(summed)
red_percentage = round(red_pixel * 100 / tot_pixel, 2)

rtg_percentage = round(red_pixel * 100 / green_pixel , 2)

print("Green pixels: " + str(green_pixel))
print("Red pixels: " + str(red_pixel))
print("Total pixels: " + str(tot_pixel))
print("Percentage of green pixels: " + str(green_percentage) + "%")
print("Percentage of red pixels: " + str(red_percentage) + "%")
print("Percentage of red to green: " + str(rtg_percentage) + "%")


plt.imshow(green)

plt.show()
plt.imshow(red)
```
